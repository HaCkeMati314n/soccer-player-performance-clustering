```python
#=========
# Summary:
#=========
# Collect the ManCity Players data and organize it in Dictionaries
#-----------------------------------------------------------------


def playerMetaDict(record):
  playerDict = {}
  try:
    playerDict['id'] = int(record.get('optaId'))
  except TypeError:
    playerDict['id'] = None
  playerDict['player'] = record.get('name')
  playerDict['number'] = int(record.get('number'))
  playerDict['position'] = record.get('position')
  return playerDict


playerMeta = []
if data.get('description')[0:3] == 'MCI':
  for record in data.get('homePlayers'):
    playerMeta.append(playerMetaDict(record))
else:
  for record in data.get('awayPlayers'):
    playerMeta.append(playerMetaDict(record))


for dict_ in playerMeta:
  print(dict_)
```

## <i>Feature Engineering
---
Here, I'll start to define the set of features for the analysis, which are <mark>five performance metrics</mark>: Playing time, Distance covered, Shots taken, Tackles won and Completed passes.</i>

#### <b><i>Feature 1: Playing time</b>
---
Playing time is a soccer performance metric that measures the amount of time a player spends on the field during a match. It is expressed here in seconds.</i>

```python
# Load the Lineup data file and store it in a variable
with open(MCxTOT_lineUp, 'r') as f:
  data = json.load(f)

print(len(data))
print(type(data))
```

```python
# Print the variable contents
for line in data:
  print(line)
```

```python
# Return the different formations the team took during the entire game time
for formation in data[0].get('formations'):
  print(formation)
```

```python
# Return the end time of the game's first-half and recast it as a datetime object.
for event in data[0].get('events'):
  if (event.get('type') == 'Half End') and (event.get('period') == 1):
    firstHalfEnd = datetime.strptime(event.get('timestamp'), '%H:%M:%S.%f')


print(f"The 1st half ended at: {firstHalfEnd.time()}")
```

```python
#=========
# Summary:
#=========
# Return each player's different roles during the first half of the game,
# calculate the playing time per role in seconds, generate the role acronym,
# gather each player's information in a single dictionary,
# and finally put all the dictionaries in a single list.
#---------------------------------------------------------------------------


def roleAcronym(playerRole):
  if playerRole == 'Goalkeeper':
    return 'GK'
  else:
    acronym = ''
    for part in playerRole.split():
      acronym += part[0]
    return acronym


def playerID(playerName):
  for player in playerMeta:
    for namePart in playerName.split():
      if namePart in player.get('player').split():
        return player.get('id')
        break
        break


def timeStrConv(timestamp):
  return datetime.strptime(timestamp, '%H:%M:%S.%f')


perfMetrics_ls = []
for team in data:
  if team.get('team_id') == 746:
    for player in team.get('lineup'):
      if player.get('positions') and (player.get('positions')[0].get('from_period') == 1):
        posList = []
        for i in range(len(player.get('positions'))):
          if player.get('positions')[i].get('from_period') == 1:
            if player.get('positions')[i].get('to_period') == 1:
              posDuration = timeStrConv(player.get('positions')[i].get('to')) - \
                            timeStrConv(player.get('positions')[i].get('from'))
              posList.append((roleAcronym(player.get('positions')[i].get('position')),
                              player.get('positions')[i].get('position'),
                              posDuration.total_seconds()))
            else:
              posDuration = firstHalfEnd - \
                            timeStrConv(player.get('positions')[i].get('from'))
              posList.append((roleAcronym(player.get('positions')[i].get('position')),
                              player.get('positions')[i].get('position'),
                              posDuration.total_seconds()))
              break
        perfMetrics_ls.append({'id': playerID(player.get('player_name')),
                           'player': player.get('player_name'),
                           'jersey': player.get('jersey_number'),
                           'role_playingTimeSec': posList})


for player in perfMetrics_ls:
  print(player)
```

#### <b><i>Feature 2: Distance covered</b>
---
The distance covered is a soccer performance metric that measures the total length of the path that a player travels during a match.</i>

```python
#=========
# Summary:
#=========
# Open the physical-splits data file, collect and total
# the distances covered by each player, then finally
# put each result in the relevant player-dictionary.
#------------------------------------------------------


def distCovAssign(first_CSVline, second_CSVline):
  id = int((first_CSVline[0].rsplit(' ', 1)[-1])[1:-1])
  distCov = reduce(lambda x, y: x + y, map(lambda x: float(x), second_CSVline[1: second_CSVline.index('')]))
  for player in perfMetrics_ls:
    if player.get('id') == id:
      player['distCov'] = distCov
      break


with open(MCxTOT_phySpl, 'r') as f:
  reader = csv.reader(f)
  data = list(reader)
  if data[1][0][0:3] == 'MCI':
    lineIdx = 27
    while data[lineIdx]:
      distCovAssign(data[lineIdx], data[lineIdx + 1])
      lineIdx += 13
  else:
    lineIdx = -13
    while not data[lineIdx - 1]:
      distCovAssign(data[lineIdx], data[lineIdx + 1])
      lineIdx -= 13

print(pd.DataFrame(perfMetrics_ls))
```

#### <b><i>Feature 3: Shots taken</b>
---
Shots taken is a soccer performance metric that measures the number of times a player attempts to score a goal by kicking the ball towards the opponent’s goal.</i>

```python
# Load the Event Data file
with open(MCxTOT_events, 'r') as f:
  data = json.load(f)

print('Type: {}'.format(type(data)))
print('Number of records: {}'.format(len(data)))
```

```python
#=========
# Summary:
#=========
# Open the Event Data, find and collect the entire
# team's shot instances in the game's first half,
# find the totality per player, and then
# put each result in the relevant player-dictionary.
#--------------------------------------------------


for playerRec in perfMetrics_ls:
  playerRec['shotsTaken'] = 0


shotsTaken_ls = []
for event in data:
  if event.get('period') == 1 and \
     event.get('team').get('id') == 746 and \
     event.get('type').get('id') == 16:
    shotsTaken_ls.append({'player': event.get('player').get('name'), 'shotsTaken': 1})
df = pd.DataFrame(shotsTaken_ls).groupby('player').count().reset_index()


for row in df.itertuples():
  for record in perfMetrics_ls:
    for namePart in row.player.split():
      if namePart in record.get('player').split():
        record['shotsTaken'] = row.shotsTaken
        break
        break
```

```python
#=========
# Summary:
#=========
# Open the Event Data, find and collect the entire
# team's complete-pass instances in the game's first half,
# find the totality per player, and then
# put each result in the relevant player-dictionary.
#---------------------------------------------------------


for playerRec in perfMetrics_ls:
  playerRec['completedPass'] = 0


completedPass_ls = []
for event in data:
  if event.get('period') == 1 and \
     event.get('team').get('id') == 746 and \
     event.get('type').get('id') == 30:
    if event.get('pass').get('outcome') and \
       event.get('pass').get('outcome').get('id') == 76:
      completedPass_ls.append({'player': event.get('player').get('name'), 'completedPass': 1})
    elif not event.get('pass').get('outcome'):
      completedPass_ls.append({'player': event.get('player').get('name'), 'completedPass': 1})
df = pd.DataFrame(completedPass_ls).groupby('player').count().reset_index()


for row in df.itertuples():
  for record in perfMetrics_ls:
    for namePart in row.player.split():
      if namePart in record.get('player').split():
        record['completedPass'] = row.completedPass
        break
        break
```

## <i>K-Means Cluster Analysis</i>
---
<i>After defining the relevant features for the analysis, my steps will be as follows:
1. Prepare the set of features.
2. Standardize the features. Since they may have different scales, it's crucial to<br> standardize them to ensure that each has equal weight in the analysis.
3. Assign the weights to the standardized features.
4. Choose the number of clusters to use in the analysis using the elbow method.
5. Run the analysis using the clustering algorithm K-Means from scikit-learn on the standardized, weighted features.</i>

```python
# Create a new dataframe that includes the features
features = []
for playerDict in perfMetrics_ls:
  features.append({'player': playerDict.get('player'),
                   'jersey': playerDict.get('jersey'),
                   'role': playerDict.get('role_playingTimeSec')[0][0],
                   'playingTimeSec': playerDict.get('role_playingTimeSec')[0][2],
                   'distCov': playerDict.get('distCov'),
                   'shotsTaken': playerDict.get('shotsTaken'),
                   'tacklesWon': playerDict.get('tacklesWon'),
                   'completedPass': playerDict.get('completedPass')})

features = pd.DataFrame(features)
print(features)
```

```python
# Create a StandardScaler object
scaler = StandardScaler()

# Standardize the features
features[['playingTimeSec_std',
          'distCov_std',
          'shotsTaken_std',
          'tacklesWon_std',
          'completedPass_std']] = scaler.fit_transform(features.iloc[:, 3:8])

print(features)
```

```python
# Define the weighing matrix
weighingMatrix = {
    'GK': [0.2, 0.25, 0.05, 0.4, 0.1],
    'LB': [0.15, 0.2, 0.05, 0.35, 0.25],
    'RCB': [0.15, 0.15, 0.05, 0.35, 0.3],
    'LCB': [0.15, 0.15, 0.05, 0.35, 0.3],
    'LW': [0.2, 0.25, 0.25, 0.1, 0.2],
    'RW': [0.2, 0.25, 0.25, 0.1, 0.2],
    'LCM': [0.2, 0.15, 0.1, 0.2, 0.35],
    'CF': [0.25, 0.2, 0.35, 0.05, 0.15],
    'CDM': [0.2, 0.1, 0.05, 0.3, 0.35],
    'RB': [0.15, 0.15, 0.05, 0.35, 0.3],
    'RCM': [0.2, 0.15, 0.1, 0.2, 0.35]
}

# Assign the relevant set of weights to each role
stdWgh_ls = []
for row in features.itertuples():
  stdWgh_ls.append(features[['playingTimeSec_std',
                             'distCov_std',
                             'shotsTaken_std',
                             'tacklesWon_std',
                             'completedPass_std']].iloc[row.Index] * weighingMatrix.get(row.role))

# Add the weighted, standardized features to the original dataframe
features[['playingTimeSec_stdWgh',
          'distCov_stdWgh',
          'shotsTaken_stdWgh',
          'tacklesWon_stdWgh',
          'completedPass_stdWgh']] = stdWgh_ls

print(features)
```

```python
# a mapping dictionary for axes labels
axLabel = {'playingTimeSec_stdWgh': 'Playing time',
           'distCov_stdWgh': 'Distance covered',
           'tacklesWon_stdWgh': 'Tackles won',
           'completedPass_stdWgh': 'Completed passes',
           'shotsTaken_stdWgh': 'Shots taken'}

# Generate unique features combinations of length 2
combin_2 = list(combinations(features.iloc[:, 13:].columns, 2))

# Print the features combinations
for combin in combin_2:
    print(combin)
```

```python
# Create a figure and subplots
fig, axs = plt.subplots(5, 2, figsize=(11, 22))

# Plots counter
n = 0

# Iterate through the axes objects and assign a scatter subplot to each
for i in range(5):
  for j in range(2):
    # Plot a scatter plot
    sns.scatterplot(ax=axs[i][j],
                    x=features.iloc[:, 13:][combin_2[n][1]],
                    y=features.iloc[:, 13:][combin_2[n][0]],
                    s=100)
    # Customize the subplot
    axs[i][j].set_xlabel(axLabel.get(combin_2[n][1]))
    axs[i][j].set_ylabel(axLabel.get(combin_2[n][0]))
    n += 1
# Adjust the subplots spacing and display them
plt.tight_layout()
plt.show()
```

```python
# Determine the optimal "k" number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(features.iloc[:, 13:])
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-Cluster Sum of Squares')
plt.show()
```

<i>It seems that the optimum number of clusters <b>K=5</b></i>
<br><br>

```python
# Set the number of clusters
n_clusters = 5

# Create a KMeans object
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)

# Fit the KMeans object to the standardized, weighted features
kmeans.fit(features.iloc[:, 13:])
```

```python
# Create a mapping dictionary for custom labels
labelMap = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E'}

# Assign each player to a cluster
cluster_labels = kmeans.predict(features.iloc[:, 13:])

# Map the cluster labels to custom labels
custom_labels = [labelMap[label] for label in cluster_labels]

# Add the custom labels to the original DataFrame
features = features.assign(cluster=custom_labels)

print(features)
```

```python
# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plots counter
n=0

# Iterate through the axes objects and assign a boxplot to each
for i in range(2):
  for j in range(2):
    n+=1
    # Plot a boxplot plot
    sns.boxplot(ax=axs[i][j],
                y=list(axLabel.keys())[n],
                order=labelMap.values(),
                x='cluster',
                palette='Set1',
                data=features)
    axs[i][j].set_xlabel('Cluster')
    axs[i][j].set_ylabel(list(axLabel.values())[n])

# Adjust the subplots spacing and display them
plt.tight_layout()
plt.show()
```

```python
# Calculate each cluster mean value
cluster_means = features.groupby('cluster')[list(axLabel.keys())].mean()
print(cluster_means)
```
